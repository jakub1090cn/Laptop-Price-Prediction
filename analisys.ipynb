# Laptop Price Prediction #
### Presentation of Results ###

The notebook presents the evaluation of models defined in the training.py file. The metrics used for assessment are R-squared (R2), Mean Squared Error (MSE), and Root Mean Squared Error (RMSE). Each of the trained models is imported, and then predictions are made on the test dataset.

```python
from sklearn.metrics import r2_score, mean_squared_error, root_mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
import joblib
import pandas as pd
```

### Reading data ###

```python
data_to_read = ['final_X_test.csv', 'final_X_train.csv', 'final_y_test.csv', 'final_y_train.csv']
data_frames = {}

for data in data_to_read:
    data_frames[data.split('.')[0]] = pd.read_csv(f'eval_data/{data}')

final_X_test = data_frames['final_X_test']
final_X_train = data_frames['final_X_train']
final_y_test = data_frames['final_y_test']
final_y_train = data_frames['final_y_train']
```

### Lasso Regression Model ###

```python
lasso = joblib.load('models/lasso_model.joblib')

lasso_pred = lasso.predict(final_X_test)
print(f'Coefficients:')
lasso.coef_
```

### Ridge Regression Model ###

```python
ridge = joblib.load('models/ridge_model.joblib')

ridge_pred = ridge.predict(final_X_test)
print(f'Coefficients:')
ridge.coef_
```

### Linear Regression Model ###

```python
linear = joblib.load('models/linear_model.joblib')

linear_pred = linear.predict(final_X_test)
print(f'Coefficients:')
linear.coef_
```

### Polynomial Regression Model ###

```python
poly = joblib.load('models/poly_model.joblib')

poly_features = PolynomialFeatures(degree=2)
X_train_poly = poly_features.fit_transform(final_X_train)
X_test_poly = poly_features.transform(final_X_test)
poly_pred = poly.predict(X_test_poly)
print(f'Coefficients:')
poly.coef_
```

### Tree Regression Model ###

```python
tree = joblib.load('models/tree_model.joblib')

tree_pred = tree.predict(final_X_test)

plt.figure(figsize=(20,10))
plot_tree(tree, filled=True, feature_names=final_X_test.columns, max_depth=3, fontsize=10)
plt.show()
```

### Model evaluation ###

```python
model_names = ['Linear', 'Lasso', 'Ridge', 'Polynomial', 'Decision Tree']
predictions = [linear_pred, lasso_pred, ridge_pred, poly_pred, tree_pred]
results = []

for model_name, pred in zip(model_names, predictions):
    mse = mean_squared_error(final_y_test, pred)
    r2 = r2_score(final_y_test, pred)
    r_mse = root_mean_squared_error(final_y_test, pred)
    results.append({"Model": model_name, "MSE": mse, "R2": r2, "RMSE": r_mse})

results_df = pd.DataFrame(results)
results_df[['MSE', 'R2', 'RMSE']] = results_df[['MSE', 'R2', 'RMSE']].map(lambda x: f'{x:.4f}')
```

```python
results_df
```

The Polynomial model displays an enormous value of the RMSE score and an R2 score value below 0, which is unacceptable. I will investigate whether the model is overfitted or underfitted.

```python
y_train_pred_poly = poly.predict(X_train_poly)
train_r2 = r2_score(final_y_train, y_train_pred_poly)
train_rmse = root_mean_squared_error(final_y_train, y_train_pred_poly)

test_r2 = r2_score(final_y_test, poly_pred)
test_rmse = root_mean_squared_error(final_y_test, poly_pred)

print(f"Training R^2: {train_r2}, Training RMSE: {train_rmse}")
print(f"Test R^2: {test_r2}, Test RMSE: {test_rmse}")
```

The model has learned the training data very well, but it struggles with the test data. Therefore, it is an overfitted model.

The models achieved similar R2 values around 68, which is not a very good result. Among all the models, the Ridge Regression model performed the best.
